````python
from lsst.daf.butler import Butler
````

````python
# ── Butler repository and DRP collection ──────────────────────────────────────
REPO = "main"
skymapName = "lsst_cells_v2"
instrument = "LSSTCam"
````

````python
butler = Butler(REPO, instrument=instrument)
````

````python
def find_datasets(butler, dataset, collectionPattern="*"):
    registry = butler.registry
    collections = []

    for collection in registry.queryCollections(collectionPattern):
        try:
            summary = registry.getCollectionSummary(collection)
            if dataset in summary.dataset_types:
                collections.append(collection)
        except Exception:
            # Certaines collections peuvent ne pas avoir de summary
            continue

    collections.sort()

    for c in collections:
        print(c)
````

````python
def find_datasets_opt(butler, dataset, collectionPattern="*"):
    registry = butler.registry

    collections = set()

    for ref in registry.queryDatasets(dataset, collections=collectionPattern, findFirst=True):
        collections.add(ref.run)

    for c in sorted(collections):
        print(c)
````

````python
find_datasets(butler, "preliminary_visit_image", collectionPattern="LSSTCam/prompt/output*")
````
